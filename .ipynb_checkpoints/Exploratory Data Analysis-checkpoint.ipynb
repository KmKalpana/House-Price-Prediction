```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
##Display all the columns of the dataframe.
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')

#print the shape of dataset with rows and columns.
print(dataset.shape)
```

```python
dataset.head()
```

# MISSING  Value

```python
##Here We will check the percentage of NAN values present in each feature.
features_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in features_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean(),3),'% missing values')
```

# Relationship between Missing values and Sale price.

```python
for feature in features_with_nan:
    data = dataset.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
print("Id of Houses {}".format(len(dataset.Id)))
```

# Numerical Variables

```python
#List Of numerical values.
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes !='O']
print('Number of Numberical variables: ',len(numerical_features))
```

```python
# visualise the numerical variables
dataset[numerical_features].head()
```

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
# Explore the content of these year variables
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
## analyze the temporal Datetime variables
## will check the whether there is a relation between year the house is sold and the sales price.

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs YearSold')
```

```python
## will compare the difference between All years feature with SalePrice

for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        # will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]
        #plt.bar(data[feature],data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## Numerical variables are usually of 2 type
## 1. Continous variable and 2. Discrete Variables

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
# Lets Find the realtionship between them and Sale PRice

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```
